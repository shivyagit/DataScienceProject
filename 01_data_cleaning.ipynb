```python
import pandas as pd

# Load dataset
df = pd.read_csv('../dataset/daily_household_transactions.csv')

# Display basic info
df.info()

# Show missing values
print(df.isnull().sum())

# Fix chained assignment warning by reassigning the filled columns directly
df['Subcategory'] = df['Subcategory'].fillna('Unknown')
df['Note'] = df['Note'].fillna('No Description')

# Convert 'Date' column safely
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Drop rows where critical data is missing
df.dropna(subset=['Date', 'Amount'], inplace=True)

# Convert Amount to float
df['Amount'] = df['Amount'].astype(float)

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Final data check
print("\nCleaned Data Info:")
df.info()

# Save cleaned data
df.to_csv('../dataset/cleaned_transactions.csv', index=False)

# Preview data
df.head()
```
