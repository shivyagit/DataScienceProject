```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../dataset/cleaned_transactions.csv')
df['Date'] = pd.to_datetime(df['Date'])

# Pivot table: Date x Category
pivot = df.pivot_table(index='Date', columns='Category', values='Amount', aggfunc='sum', fill_value=0)

# Correlation matrix
corr = pivot.corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation between Categories")
plt.show()
```
